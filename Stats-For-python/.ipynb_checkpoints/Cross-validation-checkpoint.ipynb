# Cross Validation
Used in ML to compare models. Thie requires the data to be divede into a training set which will be used to feed the model.

The testing set is used to evaluate the models prediction error. 

So instead of concentrating on in-sample error, as done with classical techniques, this method looks at out-of-sample prediction error. 

Models and accounts look better by over fitting the data they are trained on. x

Instead they need to in some sense understand something about the world. 

```python
import math
```

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as pp
%matplotlib inline
```

```python
import statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Loading gapminder data for year 1985(LiveAid)
gapminder = pd.read_csv('data/gapminder.csv')
```

```python
gdata = gapminder.query('year == 1985')
```

```python
# Refactored plotting so that it works on random datasets
def sc(data):
    return  1e-6 * data.population, data.region.map({'Africa': 'skyblue', 'Europe': 'gold', 
                                                     'America': 'palegreen', 'Asia': 'coral'})

def plotdata(data=gdata):
    size, colors = sc(data)
    data.plot.scatter('age5_surviving', 'babies_per_woman', 
                      c=colors, s=size, linewidths=0.5, edgecolor='k', alpha=0.5)

def plotfit(fit, data=gdata):
    plotdata(data)
    
    size, colors = sc(data)
    pp.scatter(data.age5_surviving, fit.predict(data), 
                      c=colors, s=30, linewidths=0.5, edgecolor='k', marker='D')
```

```python
# model formulas
f_groupmeans = 'babies_per_woman ~ -1 + region'
f_surviving = 'babies_per_woman ~ -1 + region + age5_surviving'
f_srp = ('babies_per_woman ~ -1 + region + age5_surviving'
         '+ age5_surviving:region - age5_surviving + population')
```

```python
# to divide up my data i first shuffle it with a pandas data frame method .sample
# the resulting data set should be the same length of the original.
# there is no replacement this time. 
shuffled = gdata.sample(len(gdata))
```

```python
# use iloc which goes by row number to divide the rows in two
# // for integer division and it shows i have a proper index
training, testing = shuffled.iloc[:len(gdata)//2], shuffled.iloc[len(gdata)//2:]
```

```python
# i will now fit the model to the training data using one of my formulas
surviving = smf.ols(f_surviving, data=training).fit()
```

```python
# plotting the results
plotfit(surviving, training)
```

```python
# the in-sample in square is .85
surviving.mse_resid
```

```python
# i will now try out the model on the testing data.
# plotfit contains a call to stats model predict, i will generate a moderate response for this particular-
# data set.
plotfit(surviving,testing)

# i see that the data points are different but the fit is comparable.
```

```python
# to obtain the mean square i need to do a little work.
# i first compute the residuals.
# apply modern prediction to the testing data set.
# subtract the observe data.
resid = surviving.predict(testing) - testing.babies_per_woman

# now i compute the number of degrees of freedom, which is the length of the data
# minus the number of parameters in the model, minus 1 since the number of parameters doesnt count-
# the constant, but i should.
df = len(testing) - surviving.df_model - 1

# finally i square up the residuals, sum them, and divide by the number of degrees of freedom.
np.sum(resid**2) / df
```

```python
# the mean square on the testing data is a little higher since now there is no overfitting.

# i will not automate this procedure and use it on the three models.
# since the results depend on the shuffling and are rather noisy, i'll repeat-
# a few times and take the average. 
```

```python
# i will make a function that is generic in terms of the formula, 
# of the response variable that i am looing at, and of the number of repititions.
def cross_validate(data, formula, response, n=100):
    ret = []
    
    for i in range(n):
        shuffled = data.sample(len(data))
        
        training, testing = shuffled.iloc[:len(data)//2], shuffled.iloc[len(data)//2:]
        
        trained = smf.ols(formula, data=training).fit()
        
        resid = trained.predict(testing) - testing[response]
        df = len(testing) - trained.df_model - 1
        mse = np.sum(resid**2) / df
        
        ret.append(mse)
    return np.mean(ret)
```

```python
# first on the simplest model
# the response variable is always babies_per_woman
cross_validate(gdata, f_groupmeans, 'babies_per_woman')
```

```python
cross_validate(gdata, f_surviving, 'babies_per_woman')
```

```python
# finally on the complicated one
cross_validate(gdata, f_srp, 'babies_per_woman')
```
