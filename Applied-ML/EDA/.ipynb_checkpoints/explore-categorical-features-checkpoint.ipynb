### Read in data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

titanic = pd.read_csv('../data/titanic.csv')
titanic.head()
```

```python
# drop all continuous features
cont_feat = ['PassengerId', 'Pclass', 'Name', 'Age', 'SibSp', 'Parch', 'Fare']
titanic.drop(cont_feat, axis=1, inplace=True)
```

```python
# i can see that my dataset is more cleaned up and i only have the-
# categorical features i want to explore and my target variable 'Survived'
titanic.head()
```

### Explore categorical features
Explore Sex, Cabin, and Embarked

```python
# i will start by exploring my data at a very high level. 
titanic.info()

# notice that Cabin has 204 non-null features and Embarked has 889 non-null features.
# that tells me that i have a lot of missing values for cambin and just a couple for embarked.
```

```python
# recall that when looking at missing data i looked at the corellation between missingness-
# and the values of the other features, to determine if it was missing in a systamatic way.
titanic.groupby(titanic['Cabin'].isnull()).mean()

# i want to call out 2 things here:
# this is only returning the survived field because it is the only feature with a number
# when it says cabin = True it means its missing
# false meaning it is not missing.

# essentially what the table is telling me is there is a really dramatic split-
# where 66%  of the people who have non-missing cabin values survived.
# while less than 30% of those who had a missing cabin value survived.

# its very clear that cabin is not missing at random, its actually a very strong indicator-
# whether somebody survived or not.

# an easy hypothesis here is that people without an assigned cabin literally dont have a cabin
```

```python
# since the missing values have some value, i want to capture this.
# i will handle this missing data by creating a binary indicator that-
# says whether a person had a cabin or not.

# the np.where method works like a conditional statement.
titanic['Cabin_ind'] = np.where(titanic['Cabin'].isnull(), 0, 1)
titanic.head(10)
```

### Plot categorical features
What i really want to understand in this section is the relationship between the different levels of my three categorical variables and the survival rate.

This will tell me whether for instance, women were more likely to survive than men.

This will give me an idea of which features are useful and which are not.

```python
# here i will loop through my categorical features cabin_ind, sex, embarked.
# then i will create a new plot for each index for the items in my list 
# finally i create the categorical plot where i use the feature name, plotting that against survived-
# on the Y-axis.
for i, col in enumerate(['Cabin_ind', 'Sex', 'Embarked']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=titanic, kind='point', aspect=2)
    
# first plot: people without cabins had a 30% survival rate, and those who did have-
# cabins were around 66%.

# second plot: more than 70% of the women survived, while only 20% of men survived.
# its clear that this has really powerful splitting power.

# third plot: looking at the embarkation location. C = Cherbourg, Q = Queenstown, S = South Hampton
# i can see some pretty clear separation power here. its unlikely that where they boarded caused them to survive-
# or not. more than likely this plot is correlated with other features that are already being accounted-
# for in the data.

# for instance, perhaps a higher ratio of men boarded in south hampton. i can explore this hypothesis-
# using pivot tables.
```

```python
# pivot tables are good for exploring the relationship between multiple variables. 
# i pass in the aggregation function for count because i want to look at the-
# distribution of where people boarded based on gender
titanic.pivot_table('Survived', index='Sex', columns='Embarked', aggfunc='count')

# this is telling me that 95% of people that boarded in Cherbourg were male, and 73% female.
```

```python
# next i will look at the relationship between port and whether they had cabins or not.
# my hypothesis would be that more people more people that boarded in Cherbourg had cabins-
# and thats why i saw a higher survival rate.
titanic.pivot_table('Survived', index='Cabin_ind', columns='Embarked', aggfunc='count')

# here i can see that for Queenstown and South Hampton there are drastically more people-
# without cabins than with cabins.

# in this section i learned that cabin indicator and sex have a very strong correlation with survival-
# and they could be really useful features in a model. 

# i also learned that embarked is not providing much info that isnt already covered by the other features-
# in the model.
```
